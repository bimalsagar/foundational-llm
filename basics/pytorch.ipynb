### this notebook is to understand and revise operations and functionality in pytorch such as model development and training

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
torch.manual_seed(17)

X = torch.linspace(-10, 10, 100).unsqueeze(1)

y = 2*X + 2

print(X.shape)
print(y.shape)
```

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)
```

```python
model = LinearRegressionModel()
```

```python
for name, param in model.named_parameters():
    print(name, param.data)
```

```python
cost = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 200

for epoch in range(epochs):

    model.train()

    predictions = model(X)
    loss = cost(predictions, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}')
```

```python
model.eval()

with torch.no_grad():
    test = torch.tensor([[4.0]])
    prediction = model(test)
    print(f'Prediction for input 4.0: {prediction.item()}')
```
